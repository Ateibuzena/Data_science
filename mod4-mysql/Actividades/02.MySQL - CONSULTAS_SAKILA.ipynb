```python
import pandas as pd
import mysql.connector
from datetime import datetime, timedelta
import random
```

```python
# ESTABLECER CONEXIÓN CON LA BASE DE DATOS DE MYSQL
# PARA GUARDAR LA CONEXIÓN Y EL CURSOR, HAY QUE IGUALAR LA FUNCIÓN A DOS VARIABLES
def conexion():
    db = mysql.connector.connect(host     = "localhost",
                             user     = "root", # poned vuestro usuario
                             password = "2JTovldO", # poned vuestra contraseña
                             database = None)
    cursor = db.cursor()
    return db, cursor

# CERRAR LA CONEXIÓN
def cerrar_conexion(cursor, db):
    cursor.fetchall() # se utiliza para recuperar todas las filas del conjunto de resultados después de ejecutar una consulta en una base de datos. Esto es útil cuando deseas trabajar con los resultados de una consulta en tu programa Python
    cursor.close()
    db.close()

# EJECUTAR QUERYS DE CONSULTAS:
def ejecutar_consulta(database, query):
    db, cursor = conexion()
    cursor.execute(f"USE {database};")

    # Ejecutar la consulta
    cursor.execute(query)
    # Obtener los resultados, si es una consulta SELECT
    resultados = cursor.fetchall()
    # Hacer commit para guardar los cambios (en caso de INSERT, UPDATE, DELETE, etc.)
    db.commit()
    # Cerrar la conexión
    cerrar_conexion(cursor, db)
    return resultados
```

### EJERCICIO1: LENGTH()

1. Escribe una query para encontrar todos los clientes cuyo nombre tenga más de 7 caracteres.

```python
query1 = """SELECT customer.first_name 
        FROM customer
        WHERE length(customer.first_name)> 7;
        """
ejecutar_consulta("sakila", query1)
```

3. Escribe una query para calcular la longitud promedio de los nombres de los clientes.

4. Escribe una query para encontrar la longitud del nombre del actor más largo.

```python
query4 = """SELECT actor.first_name, length(actor.first_name) as longitud_nombre
        FROM actor
        GROUP BY actor.first_name
        ORDER BY longitud_nombre DESC LIMIT 1;
        """
ejecutar_consulta("sakila", query4)
```
