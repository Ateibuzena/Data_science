```python
def obtener_links(comunidad):
    campo_input = driver.find_element(By.CSS_SELECTOR, '#searchboxinput')

    campo_input.send_keys(f"campings {comunidad}")
    driver.find_element(By.CSS_SELECTOR, '#searchbox-searchbutton').click()
    sleep(5)
    enlaces = driver.find_elements(By.CLASS_NAME, 'hfpxzc')
    for b in enlaces:
        links.append(b.get_attribute("href"))
    driver.find_element(By.CSS_SELECTOR, '#searchbox > div.lSDxNd > button').click()
    return links
```

```python
options = Options()
options.add_experimental_option('excludeSwitches', ['enable-automation'])
options.add_experimental_option('useAutomationExtension', False)
#PARA QUE NO TE BANEEN, A PARTIR DE X REQUESTS TE BANEAN
from fake_useragent import UserAgent
user = UserAgent().random
# ahora añadiremos nuestro usuario al objeto de opciones
options.add_argument(f'user-agent={user}')
driver = webdriver.Chrome(chrome_options=options, executable_path='../5.Web_scraping/chromedriver.exe')
# Abrir una página web
driver.get("https://www.google.es/maps/@40.2085,-3.713,6z?entry=ttu")
driver.find_element(by = By.CSS_SELECTOR, value = "#yDmH0d > c-wiz > div > div > div > div.NIoIEf > div.G4njw > div.AIC7ge > div.CxJub > div.VtwTSb > form:nth-child(2) > div > div > button > span").click()

sleep(5)

comunidades = ["Andalucia", "Aragon", "Asturias", "Cantabria", "Castilla la Mancha", "Castilla y León", "Cataluña", "Extremadura", "Galicia", "Islas Baleares", "Islas Canarias", "La Rioja", "Madrid", "Murcia", "Navarra", "Pais Vasco", "Valencia"]
direcciones = []
webs = []
telefonos = []
geolocalización = []
cosas = []
links = []
latitudes = []
longitudes = []
```

```python
for comunidad in comunidades:
    obtener_links(comunidad)
```

```python
for link in links:
    print(link)
    driver.get(link)
    
    soup = BeautifulSoup(driver.page_source, "html.parser")
    cositas = soup.find_all("div", class_="AeaXub")
    sleep(3)
    for c in cositas:
        
        cosas.append(c.text)
    
    direcciones.append(cosas[0])
    webs.append(cosas[1])
    telefonos.append(cosas[2])
    cosas = []
```

```python
len(links), len(direcciones), len(telefonos), len(webs)
```

```python
for n in range(len(direcciones)):
    try:
        direcciones[n] = direcciones[n].replace(" s/n,", "")
        direcciones[n] = direcciones[n].replace(" S/n,", "")
        direcciones[n] = direcciones[n].replace(" S/N,", "")
        direcciones[n] = direcciones[n].replace("Crta.", "Carretera")
        direcciones[n] = direcciones[n].replace("Cam.", "Camino")
    except:
        pass
```

```python
for n in range(len(direcciones)):
    geolocator = Nominatim(user_agent="tu_aplicacion")
    direccion = direcciones[n]
    
    location = geolocator.geocode(direccion)
    if location:
        latitud = location.latitude
        longitud = location.longitude
        lista = [latitud, longitud]
        geolocalización.append(lista)
        latitudes.append(latitud)
        longitudes.append(longitud)
        lista = []
        print(f"Latitud: {latitud}, Longitud: {longitud}")
        world_map = folium.Map(location = [latitud,longitud], zoom_start = 20)

        world_map
    
    else:
        latitudes.append(np.nan)
        longitudes.append(np.nan)
        geolocalización.append(np.nan)
    
```

```python
len(latitudes),len(geolocalización)
```

```python
data = {"direcciones":direcciones,
        "webs": webs,
        "telefonos": telefonos,
        "geolocalización": geolocalización,
        "latitud": latitudes,
        "longitud": longitudes}

df_campings = pd.DataFrame(data)
```

```python
# Eliminar '.com' y crear la columna 'nombre'
df_campings['camping'] = df_campings['webs'].str.rstrip('.com')
```

```python
#mover la columna camping a la posicion 1:
df_campings.insert(0, 'camping', df_campings.pop('camping'))
```

```python
# Guardar en un archivo Excel
df_campings.to_excel('Campings_espana.xlsx', index=False)  # El argumento index=False evita guardar el índice del DataFrame como una columna adicional
```

```python
latitude = 39.924432
longitude = -7.9092288
# Inicializa el mapa ubicado en San Francisco
sf_map = folium.Map(location = [latitude, longitude], zoom_start = 5)

# Muestra el mapa
sf_map
# Inicializamos un FeatureGroup() para los incidentes en el DataFrame
incidents = folium.map.FeatureGroup()

# Recorre los 100 crímenes y agrega a cada uno al FeatureGroup() de incidentes
# la columna Y y X son las coordenadas, latitud y longitud respectivamente

for lat, lng, label in zip(df_campings.latitud, df_campings.longitud, df_campings.camping):
    try:
        incidents.add_child(folium.Marker(location = [lat, lng],
                                      popup    = label))
    except:
        pass

# Agrega incidentes al mapa
sf_map.add_child(incidents)

sf_map
```
