### Ejercicio 15:
- Encontrar los 5 mejores pokémons con respecto a la columna "Total", del tipo "Electric", "Fire", "Water" y comparar su salud, ataque, defensa y velocidad en un mapa de calor.

```python
df_filtrado1 = df[(df["Type 1"].isin(["Fire"]))].sort_values("Total", ascending= False)[0:5]
df_filtrado2 = df[(df["Type 1"].isin(["Electric"]))].sort_values("Total", ascending= False)[0:5]
df_filtrado3 = df[(df["Type 1"].isin(["Water"]))].sort_values("Total", ascending= False)[0:5]
df_filtrado4 = df[(df["Type 2"].isin(["Electric"]))].sort_values("Total", ascending= False)[0:5]
df_filtrado5 = df[(df["Type 2"].isin(["Fire"]))].sort_values("Total", ascending= False)[0:5]
df_filtrado6 = df[(df["Type 2"].isin(["Water"]))].sort_values("Total", ascending= False)[0:5]
dataframes = [df_filtrado1, df_filtrado2, df_filtrado3, df_filtrado4, df_filtrado5, df_filtrado6]

df_final = pd.concat(dataframes, ignore_index=True)
los5mejores = df_final.sort_values("Total", ascending=False)[0:5]
```

### Ejercicio 16:
- Crear un poligono para 5 pokemons con sus caracteristicas. (Adapta el código del notebook de teoria)

```python
los5mejores.reset_index(inplace=True)
```

```python
los5mejores
```

```python
labels = np.array(los5mejores.columns[5:11])

pattern = los5mejores.iloc[0, 5:11]

angles = np.linspace(0, 2*np.pi, len(labels), endpoint = False)
print(angles)

# Esto es para añadir el último lado del polígono:
pattern = np.concatenate((pattern, [pattern[0]]))
angles = np.concatenate((angles, [angles[0]]))


fig = plt.figure()
ax = fig.add_subplot(polar = True)

ax.plot(angles, pattern, "o-", linewidth = 4)           # Lineas en azul
ax.fill(angles, pattern, alpha = 1, color = "yellow")   # Relleno amarillo
ax.set_thetagrids(angles[:-1] * 180/np.pi, labels)      # Malla
plt.title(los5mejores.Name[0], fontsize=20, color = "cyan")
plt.show()
```

### Ejercicio 17:
- Usando FacetGrid obtén un KDE plot del ataque para legendarios y no legendarios.
- Repite lo mismo, ahora separándolos también por generaciones en cada fila.

```python
g = sns.FacetGrid(df, col = "Legendary")
g.map(sns.kdeplot, "Attack")
plt.show(); #punto y coma para que no salga el objeto
```

```python
g = sns.FacetGrid(df, col = "Legendary", row= "Generation")
g.map(sns.kdeplot, "Attack")
plt.show(); #punto y coma para que no salga el objeto

#el kdeplot, el eje y indica la probabilidad de que una observación caiga dentro de un cierto rango de valores.
#en este caso la probabilidad de que cogiendo un pokemon al azar tenga ese ataque
```
