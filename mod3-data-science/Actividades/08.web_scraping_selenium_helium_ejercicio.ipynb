```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
```

```python
browser = webdriver.Chrome('../5.Web_scraping/chromedriver.exe')
browser.get("https://www.20minutos.es/ciencia/")
browser.find_element(by = By.CSS_SELECTOR, value = "#didomi-notice-agree-button").click()
# Obtener el identificador de la ventana actual (la página inicial)
# Esperar hasta que los enlaces estén presentes en la página
```

```python
enlaces = browser.find_elements(By.TAG_NAME, "a")   #sin ese
```

```python
lista_link = []

# Filtrar los enlaces que contengan "5181"
for enlace in enlaces:
    if enlace.get_attribute("href") is not None:
        link = enlace.get_attribute("href")
        if "5181" in link and link is not None:
            lista_link.append(link)
```

```python
horas = []
fechas = []
autores = []
categorias = []
textos = []
for a in lista_link:
    browser.get(a)
    fechayhora = browser.find_element(By.CLASS_NAME, "article-date")
    texto = fechayhora.text.split("-")
    horas.append(texto[1])
    fechas.append(texto[0])
    author = browser.find_element(By.CLASS_NAME, "article-author")
    autor = author.text
    autores.append(autor)
    category = browser.find_element(By.CSS_SELECTOR, "#block_1 > section > span.tooltip.tooltip-article-info")
    categorias.append(category.text)
    texto = browser.find_element(By.CSS_SELECTOR, '#block_1 > div.container-grid-box > section > article > div.article-text.__hnads_custom_targetings')
    textos.append(texto.text.replace("\n", " "))
```

```python
articulos = []
horas = []
fechas = []
autores = []
categorias = []
textos = []
lista_link = []
```

```python
options = Options()
options.add_experimental_option('excludeSwitches', ['enable-automation'])
options.add_experimental_option('useAutomationExtension', False)
#PARA QUE NO TE BANEEN, A PARTIR DE X REQUESTS TE BANEAN
from fake_useragent import UserAgent
user = UserAgent().random
# ahora añadiremos nuestro usuario al objeto de opciones
options.add_argument(f'user-agent={user}')
amapola = webdriver.Chrome('../5.Web_scraping/chromedriver.exe')
```

```python
for pagina in paginas:

    amapola.get(pagina)
    try:
        amapola.find_element(by = By.CSS_SELECTOR, value = "#didomi-notice-agree-button").click()

    except:
        pass
    clase = amapola.find_elements(By.CLASS_NAME, 'media-content')
    for titulo in clase:
        try:
            texto = titulo.text
            articulos.append(texto.split("\n")[0])
        except:
            pass
    
    enlaces = amapola.find_elements(By.TAG_NAME, "a")
    for enlace in enlaces:
        if enlace.get_attribute("href") is not None:
            link = enlace.get_attribute("href")
            if "5181" in link and link is not None:
                lista_link.append(link)

    for a in set(lista_link):

        print(a)

        amapola.get(a)
        fechayhora = amapola.find_element(By.CLASS_NAME, "article-date")
        texto = fechayhora.text.split("-")
        horas.append(texto[1])
        fechas.append(texto[0])
        author = amapola.find_element(By.CLASS_NAME, "article-author")
        autor = author.text
        autores.append(autor)
        category = amapola.find_element(By.CSS_SELECTOR, "#block_1 > section > span.tooltip.tooltip-article-info")
        categorias.append(category.text)
        texto = amapola.find_element(By.CSS_SELECTOR, '#block_1 > div.container-grid-box > section > article > div.article-text.__hnads_custom_targetings')
        textos.append(texto.text.replace("\n", " "))
        sleep(1)
        
```

```python
pd.DataFrame(data = zip(articulos, fechas, horas, categorias, autores, textos), 
             columns= ["articulos", "fechas", "horas", "categorias", "autores", "textos"])
```
