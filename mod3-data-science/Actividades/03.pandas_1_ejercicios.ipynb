### Ejercicio 8:
- Eliminar las columnas: **`Tipo`**, **`Area`** y **`Region`**.

```python
# en el drop, rename, etc: axis = 1 significa columnas y axis=0 significa filas
#en los demás metodos como max, min, median, etc: axis = 1 significa filas y axis = 0 significa columna
```

Se está eliminando una columna, lo que implica una operación a lo largo del eje horizontal (a través de las columnas). Por lo tanto, la orientación es horizontal y axis=1 es apropiado en este contexto.

La clave aquí es que, a pesar de que los datos están dispuestos verticalmente en una columna, desde la perspectiva de la operación de eliminación, se está tratando con las columnas como una entidad colectiva. Por eso, axis=1 indica que la operación se realiza en la dirección de las columnas.

```python
#HAY QUE PONER AXIS = 1 PARA QUE ELIMINE LA COLUMNA HACIA ABAJO, dentro del parentesis al lado del nombre de la columna
#Si omites el argumento axis=1 al utilizar el método drop en un DataFrame, pandas podría generar un error porque necesita saber si estás tratando de eliminar una columna o una fila.
df.drop("Tipo", axis = 1, inplace = True)  #inplace = true dentro del parentesis del metodo
```

```python
#eliminar varias columnas a la vez, las columnas tienen que estar en forma de lista:
df.drop(["Area", "Region"], axis = 1, inplace = True)
```

### Ejercicio 26:
- Encontrar el año en que hubo más migración por continente y crear un dataframe con nombre de continente, el año de máxima migración y cantidad de migración.

|	  |Continente|Año |Migraciones|
|-----|----------|----|-----------|
|**0**|Asia	     |2005|42584      |
|**1**|Europe	 |1994|39231      |
|**2**|Africa	 |2010|6242       |
|**3**|Oceania	 |1992|1782       |
|**4**|America	 |2008|10190      |

```python
#nombres de continentes:
df["Continente"].unique()
```

```python
#listas vacias:
continentes = list()
años = list()
migraciones = list()

#recorremos los continentes:
for cont in df["Continente"].unique():
    filas_porcontinente = df[df["Continente"] == cont]
    #filtra el DataFrame df para obtener solo las filas donde el valor en la columna "Continente" coincide con el valor actual de cont
    filas_desde1980a2013 = filas_porcontinente[range(1980,2014)]
    # selecciona las columnas desde 1980 hasta 2013
    max_columnas = filas_desde1980a2013.max()
    #Calcula el valor máximo para cada columna seleccionada (es decir, para cada año).
    max_columnas_ordenadas = max_columnas.sort_values()
    #Ordena los valores de máximo a mínimo. Esto significa que los años estarán ordenados desde el año con el máximo valor hasta el mínimo.
    año_maximo = max_columnas_ordenadas.tail(1)
    #el año con el valor maximo está en el ultimo valor
    nombre_año = año_maximo.index[0]
    cantidad_migracion = año_maximo.values[0]

    años.append(nombre_año)
    migraciones.append(cantidad_migracion)
    continentes.append(cont)
años, migraciones, continentes

```

```python
df_migracion = pd.DataFrame()
```

```python
df_migracion["Continente"] = continentes
df_migracion["Año"] = años
df_migracion["Migraciones"] = migraciones
```
