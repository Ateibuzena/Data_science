### Ejercicio 10
- Cual es el jugador mas joven y cual es el mas viejo.

```python
# Jugadores más jóvenes
df.select("*")\
  .where(df["Age"] == (df.select("Age")\
                         .orderBy(df["Age"].asc())\
                         .collect()[0][0])
        )\
  .toPandas()
```

```python
# Jugador más viejo
df.select("*")\
  .where(df["Age"] == (df.select("Age")\
                         .orderBy(df["Age"].desc())\
                         .collect()[0][0])
        )\
  .toPandas()
```
