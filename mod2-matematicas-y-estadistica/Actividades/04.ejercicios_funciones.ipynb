### Ejercicio 4: 
Dada la siguiente funci√≥n: 

$$x^2+10x+1 = 0$$ 

- Genera la grafica de la funci√≥n, encuentra su derivada y muestra la recta tangente de la derivada en el punto x = -7.5

```python

fx = x**2 + 10*x + 1

px = -7.5
```

```python
def grafica_funcion_condatos(fx, px):
    x = Symbol("x")
    x1 = px
    y1 = fx.subs(x, x1)

    fx_derivada = Derivative(fx).doit()
    
    m = fx_derivada.subs(x, x1)

    recta_tangente = m*x - m*x1 + y1

    print(f"Punto donde calculamos la derivada: ({x1}, {y1})")
    print(f"Pendiente o resultado de la derivada en ese punto: {m}")
    print(f"Ecuacion de la recta tangente en ese punto: y = {recta_tangente}")

    valores_x = np.linspace(-20,20,100)

    y = [fx.subs(x, i) for i in valores_x]
    plt.plot(valores_x, y , color = "pink")

    y_rectatangente = [recta_tangente.subs(x,i) for i in valores_x]
    plt.plot(valores_x, y_rectatangente, color = "purple")

    plt.plot(x1, y1, color = "blue", marker = "o")

    plt.xlim(-20, 20)
    plt.grid()
    plt.show()
    
```

```python
grafica_funcion_condatos(fx, px)
```

```python
grafica_funcion_condatos(2*x**2 + 4*x - 3, 0)
```
