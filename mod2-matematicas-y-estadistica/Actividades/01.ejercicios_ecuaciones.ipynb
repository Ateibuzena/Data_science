```python
import numpy as np

import math

import matplotlib.pyplot as plt
```

```python
def ecuacion_segundogrado(a, b, c):
    # Eje X
    plt.plot(np.linspace(-10, 10, 30), [0 for i in range(30)], color = "black", alpha = 0.5) 
    # Eje Y
    plt.plot([0 for i in range(30)], np.linspace(-10, 50, 30), color = "black", alpha = 0.5) 
    x = symbols("x")
    if (b**2 - 4*a*c) < 0:
        
        x = np.linspace(-20, 20, 100)
        fx = a*x**2 + b*x + c
        plt.plot(x, fx)
        plt.axis("off")
        plt.show()
    else:
        
        fx = a*x**2 + b*x + c
        x1 = (-b + math.sqrt(b**2 - 4*a*c)) / (2 * a)
        x2 = (-b - math.sqrt(b**2 - 4*a*c)) / (2 * a)
        print(f"x1 = {x1}, x2 = {x2}")
        plt.plot(x1, fx.subs(x, x1), marker = "o", color = "red")
        plt.plot(x2, fx.subs(x, x2), marker = "o", color = "red")
        x = np.linspace(-20, 20, 100)
        fx = a*x**2 + b*x + c
        plt.plot(x, fx)
        plt.axis("off") #Eliminar ejes de grafica
        plt.show()

 
```

```python
def ecuacion_cuadratica_1(a, b, c):
    
    # x, y son np.arrays
    x = np.linspace(-10, 10, 100)
    y = a*x**2 + b*x + c
    
    # Curva
    plt.plot(x, y, color = "black")
    
    # Eje X
    plt.plot(np.linspace(-10, 10, 30), [0 for i in range(30)], color = "black", alpha = 0.5) 

    # Eje Y
    plt.plot([0 for i in range(30)], np.linspace(-10, 50, 30), color = "black", alpha = 0.5) 
    
    if b**2 - 4*a*c >= 0: 
        x1 = (-b + math.sqrt(b**2 - 4*a*c)) / (2*a)
        x2 = (-b - math.sqrt(b**2 - 4*a*c)) / (2*a)
        
        # Raíz 1
        plt.plot(x1, 0, color = "red", marker = "o")

        # Raíz 2
        plt.plot(x2, 0, color = "red", marker = "o")
                
        plt.show()
        
        return x1, x2
        
    else:
        print("Raíces imaginarias")
        plt.show()
```

### Ejercicio 8:
- Encuentra la distancia entre los dos puntos:

    - P1(-3, 6)
    - P2(5, 10)

#distancia entre dos puntos: 
$$D(P1,P2) = \sqrt{(x_2-x_1)^2 + (y_2-y_1)^2} $$

```python
def distancia_dos_puntos(P1, P2):
    x1 = P1[0]
    x2 = P2[0]
    
    y1 = P1[1]
    y2 = P2[1]
    
    distancia = sqrt((x2 - x1)**2 + (y2 - y1)**2)
    
    return distancia
```

```python
distancia_dos_puntos((0,0), (2,2))
```

### Ejercicio 9:
- Encuentra la distancia entre los dos puntos:

    - P1(-2, -3, -4)
    - P2(3, 4, 2)

𝑑(𝑃1, 𝑃2) = √(𝑥1 − 𝑥2)**2 + (𝑦1 − 𝑦2)**2 + (𝑧1 − 𝑧2)**2

```python
def d_dospuntos_3coordenadas(P1, P2):
    x1 = P1[0]
    x2 = P2[0]
    
    y1 = P1[1]
    y2 = P2[1]
    
    z1 = P1[2]
    z2 = P2[2]
    
    distancia = sqrt((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)
    
    return distancia
```

```python
d_dospuntos_3coordenadas((-2, -3, -4), (3, 4, 2))
```

```python
def d_entrepuntos_nd(P1, P2):
    radicando = 0
    for i, n in zip(P1, P2):
        radicando += (n - i)**2
        print(radicando)
    distancia = sqrt(radicando)
    return distancia
```

### Ejercicio 12:
- Crea una función para la **distancia Manhhatan** y otra para la **distancia Chebyshev**. Encuentra las distancias entre el punto A y los puntos B y C del ejercicio anterior utilizando las dos funciones.

$$Manhattan = \sum_{i = 0}^{n}\left| x_{i} - y_{i} \right|$$

```python
def distancia_manhattan(P1, P2):
    distancia = 0
    for i, n in zip(P1, P2):
        absoluto = (n - i)
        
        if absoluto < 0:
            distancia += absoluto*(-1)
        else:
            distancia += absoluto
        
    return distancia
```

```python
distancia_manhattan(P1 = (1.2, 3.4, -3, 5, 6), P2 = (11, 2/3, 1, 10, 12))
```

```python
def distancia_chebyshev(P1, P2): 
    distancias = list()
    for x, y in zip(P1, P2):
        distancias.append(abs(y - x))
    distancia = max(distancias)
    return distancia
```

```python
distancia_chebyshev(P1 = (1.2, 3.4, -3, 5, 6), P2 = (11, 2/3, 1, 10, 12))
```
