Pero el m√≥dulo es la medida entre el punto de origen y el punto extremo. Entonces ¬øPorqu√© no es la longitud de la recta que une el punto inicial con el final?:

```python
def distancia_dos_puntos(P1, P2):
    x1 = P1[0]
    x2 = P2[0]
    
    y1 = P1[1]
    y2 = P2[1]
    
    distancia = sqrt((x2 - x1)**2 + (y2 - y1)**2)
    
    return distancia
```

```python
puntofinal = (9.5, 7.5)

modulo = distancia_dos_puntos(puntoinicial, puntofinal)
```

### Producto escalar de dos vectores

Se obtiene de la suma de los productos formados por las componentes de uno y otro vector. Es decir, dados dos vectores $\overrightarrow{AB}$ y $\overrightarrow{CD}$, expresados en un mismo sistema de coordenadas:
$$\overrightarrow{AB} = x_{\overrightarrow{AB}}*\overrightarrow{i}+y_{\overrightarrow{AB}}*\overrightarrow{j}$$
$$\overrightarrow{CD} = x_{\overrightarrow{CD}}*\overrightarrow{i}+y_{\overrightarrow{CD}}*\overrightarrow{j}$$


$$\overrightarrow{AB}\cdot \overrightarrow{CD}= x_{\overrightarrow{AB}}* x_{\overrightarrow{CD}}+y_{\overrightarrow{AB}}*y_{\overrightarrow{CD}}$$


El producto escalar tambi√©n viene dado por:

$$\overrightarrow{AB}\cdot \overrightarrow{CD} = \left | \overrightarrow{AB} \right | \cdot \left | \overrightarrow{CD} \right | \cdot cos (\alpha)$$

Donde $\left | \overrightarrow{AB} \right |$ y $\left | \overrightarrow{CD} \right |$ son los m√≥dulos de sus vectores y $\alpha$ es el √°ngulo que se forma al unir ambos vectores.

$\overrightarrow{AB} = 3\widehat{i} + 4\widehat{j} $


$\overrightarrow{CD} = -2\widehat{i} + 7.5\widehat{j} $
 

```python
AB = (3, 4)
CD = (-2, 7.5)
producto_escalar = np.dot(AB,CD)
modulo_ab = np.linalg.norm(AB)
modulo_cd = np.linalg.norm(CD)
# Radianes
angulo_radianes = np.arccos(producto_escalar / (modulo_ab * modulo_cd))
#Grados
angulo_grados = np.degrees(angulo_radianes)
angulo_grados
```
